```python
import pandas as pd
import matplotlib.pyplot as plt

birds = pd.read_csv('data/birds.csv')
birds.head()
```

```python
plt.title('Max BodyMass in Centimeters')
plt.ylabel('BodyMass (CM)')
plt.xlabel('Birds')
body_mass = birds.MaxWingspan
body_mass.plot()
```

```python
# Show the birds' names correlated to their body mass

plt.title('Max BodyMass in Centimeters')
plt.ylabel('BodyMass (CM)')
plt.xlabel('Birds')
plt.xticks(rotation=45)
x = birds['Name']
y = birds['MaxBodyMass']

plt.plot(x, y)

plt.show()
```

```python
# Label the birds whose difference between max body mass and min body mass is equal to 100


plt.title('Max BodyMass in Centimeters')
plt.ylabel('BodyMass (CM)')
plt.tick_params(axis='both',which='both',labelbottom=False,bottom=False)

for i in range(len(birds)):
    x = birds['Name'][i]
    y = birds['MaxBodyMass'][i]
    plt.plot(x, y, 'bo')
    if (birds['MaxBodyMass'][i]-birds['MinBodyMass'][i]) == 2000:
        plt.text(x, y * (1 - 0.05), birds['Name'][i], fontsize=12)

plt.show()
```

```python
# bar chart
birds.plot(x='Category',
        kind='bar',
        stacked=True,
        title='Birds of Minnesota')
```

```python
# draw bar chart by birds category
category_count = birds.value_counts(birds['Category'].values, sort=True)
plt.rcParams['figure.figsize'] = [6, 12]
category_count.plot.barh()
```

```python
minBodyMass = birds['MinBodyMass']
maxBodyMass = birds['MaxBodyMass']
category = birds['Category']

plt.barh(category, minBodyMass)
plt.barh(category, maxBodyMass)

plt.show()
```
